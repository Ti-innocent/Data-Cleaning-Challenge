## Lists

Given a list of integers, write a function that multiply all the numbers in the list. The function should be written in pure python (no external library)

Examples:

Input \[1, 2, 3] Output 6

Input \[10, 2, 2] Output 40

```python
# Code your answer here
def multi(s):
    a=1
    for i in s:
        a=a*i
    return a
A=[1,2,3]
B=[10,2,2]
print("The product of list A elements is:", multi(A))
print("The product of list A elements is:", multi(B))
```

## Dictionnary

Given a dictionnary of tuples, write a function to retrieve the keys where the second element of the tuple is negative.


Examples:

Input {'A' : (10, 30), 'B': (3, 2), 'C': (10, -1)} Ouput \[C]

Input {'A' : (-1, 10), 'B': (-3, -2), 'C': (6, -10)} Ouput \[B, C]

```python
# Code your answer here
def dict1(d):
    # a=[]
    key=[i for i,j in d.items() if j[1]<0]
    return key
dic={'A' : (10, 30), 'B': (3, 2), 'C': (10, -1)}
dic1={'A' : (-1, 10), 'B': (-3, -2), 'C': (6, -10)}
print("The keys with second element of the tuple which is negative is:", dict1(dic))
print("The keys with second element of the tuple which is negative is:", dict1(dic1))
```

You are provided with a poorly formatted excel file data/dirty_data.xlsx. In order to use this data it must be standardized and cleaned appropriately. The table contains different measurements related to passenger vehicles for different countries.

To solve this task, you must:

**The clean table should have the following format with 5 columns**:

- Measure: string containing the kind of measurement 

- Vehicle Type: string containing the type of vehicle

- Country: string containing the country

- Year: int representing the year

- Value: float representing the value of the measurement


```python
import pandas as pd 
import numpy as np
#When your data are from drive
#from google.python import drive
#drive.mount("/drive")
#data=pd.read_excel("/drive/My Drive/dirty_data.xlsx")
data=pd.read_excel("dirty_data.xlsx")
```

```python
#Checking for columns name
data.columns
```

```python
#Checking for shape/dimensions and data types of columns
print(data.shape)
data.dtypes
```

```python
#Checking for null values sum in each column
data.isna().sum()
```

```python
#Renaming the first three columns
rename={"Unnamed: 0":"Measure","Unnamed: 1":"Vehicle Type","Unnamed: 2":"Country"}
```

```python
data.rename(columns=rename,inplace=True)
data
```

```python
#Transpose of data
transpose = pd.melt(data, id_vars=['Measure','Vehicle Type','Country'],
            value_vars=['1993', '1994', '1995','1996', '1997', '1998',
            '1999', '2000', '2001','2002', '2003', '2004','2005', '2006',
            '2007', '2008', '2009','2010', '2011','2012', '2013', '2014',
            '2015', '2016','2017', '2018', '2019'])
```

```python
#Transposed data and by default value_vars named "Variable"
transpose 
```

```python
# Renaming column "variable" to Year
transpose.rename(columns={"variable":"Year"} ,inplace=True)
transpose
```

```python
# Removing the rows containing ".." in the transposed data
NewData =transpose[transpose.value != '..']
NewData
```

```python
# Reseting index
data_clean = NewData.reset_index(drop=True)
data_clean
```

```python
data_clean.to_csv("data_clean.csv")
```
